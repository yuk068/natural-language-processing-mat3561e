````python
from transformers import pipeline
# 1. Tải pipeline "fill-mask"
# Pipeline này sẽ tự động tải một mô hình mặc định phù hợp (thường là một biến thể của BERT)
mask_filler = pipeline("fill-mask")
# 2. Câu đầu vào với token [MASK]
input_sentence = "Hanoi is the <mask> of Vietnam."
# 3. Thực hiện dự đoán
# top_k=5 yêu cầu mô hình trả về 5 dự đoán hàng đầu
predictions = mask_filler(input_sentence, top_k=5)
# 4. In kết quả
print(f"Câu gốc: {input_sentence}")
for pred in predictions:
  print(f"Dự đoán: '{pred['token_str']}' với độ tin cậy: {pred['score']:.4f}")
  print(f" -> Câu hoàn chỉnh: {pred['sequence']}")
````

1.  **Mô hình đã dự đoán đúng từ `capital` không?**
    Dựa trên kết quả thực thi trước đó, mô hình đã dự đoán chính xác từ `capital` là lựa chọn hàng đầu để điền vào chỗ trống `<mask>` trong câu "Hanoi is the <mask> of Vietnam." với độ tin cậy rất cao (0.9341).

    ```
    Dự đoán: ' capital' với độ tin cậy: 0.9341
     -> Câu hoàn chỉnh: Hanoi is the capital of Vietnam.
    ```

2.  **Tại sao các mô hình Encoder-only như BERT lại phù hợp cho tác vụ này?**
    Các mô hình Encoder-only như BERT rất phù hợp cho tác vụ *fill-mask* (hay còn gọi là Masked Language Modeling - MLM) vì những lý do sau:

    *   **Hiểu biết ngữ cảnh hai chiều (Bidirectional Context):** Kiến trúc encoder của BERT được thiết kế để xử lý toàn bộ câu đầu vào cùng một lúc, cho phép nó nhìn thấy cả ngữ cảnh bên trái và bên phải của một từ bị che (mask). Điều này rất quan trọng để đoán chính xác từ bị thiếu, vì ý nghĩa của một từ thường phụ thuộc vào các từ xung quanh nó. Ví dụ, trong câu "Hanoi is the <mask> of Vietnam.", BERT có thể sử dụng cả "Hanoi" (bên trái) và "of Vietnam" (bên phải) để suy luận rằng "capital" là từ phù hợp nhất.
    *   **Mô hình tiền huấn luyện trên nhiệm vụ MLM:** BERT được tiền huấn luyện đặc biệt trên nhiệm vụ Masked Language Modeling. Điều này có nghĩa là trong quá trình huấn luyện, một số từ trong câu được che đi (masked), và mô hình được yêu cầu dự đoán các từ đó. Quá trình này giúp BERT học được cách hiểu sâu sắc về mối quan hệ giữa các từ và ngữ pháp trong ngôn ngữ, từ đó rất hiệu quả khi điền vào các chỗ trống.
    *   **Biểu diễn ngữ cảnh (Contextual Embeddings):** BERT tạo ra các vector biểu diễn (embeddings) cho mỗi từ dựa trên ngữ cảnh của nó trong câu. Điều này khác với các phương pháp nhúng từ truyền thống chỉ cung cấp một vector cố định cho mỗi từ. Nhờ đó, BERT có thể phân biệt ý nghĩa của cùng một từ khi nó xuất hiện trong các ngữ cảnh khác nhau, giúp đưa ra dự đoán chính xác hơn cho các từ bị che.

````python
from transformers import pipeline
# 1. Tải pipeline "text-generation"
# Pipeline này sẽ tự động tải một mô hình phù hợp (thường là GPT-2)
generator = pipeline("text-generation")
# 2. Đoạn văn bản mồi
prompt = "The best thing about learning NLP is"
# 3. Sinh văn bản
# max_length: tổng độ dài của câu mồi và phần được sinh ra
# num_return_sequences: số lượng chuỗi kết quả muốn nhận
generated_texts = generator(prompt, max_length=50, num_return_sequences=1)
# 4. In kết quả
print(f"Câu mồi: '{prompt}'")
for text in generated_texts:
  print("Văn bản được sinh ra:")
  print(text['generated_text'])
````

Mô hình GPT-2 đã tiếp tục câu một cách mạch lạc, tạo ra một đoạn văn bản có ý nghĩa, mặc dù đôi khi có thể hơi lạc đề hoặc lặp ý. Cụ thể, mô hình đã mở rộng chủ đề về học NLP và những lợi ích của nó, sau đó chuyển sang các ý tưởng liên quan như học qua nghe nhạc hoặc đọc sách, duy trì một giọng điệu nhất quán. Tuy nhiên, do giới hạn `max_length` được đặt quá thấp (50 token) nhưng sau đó lại bị `max_new_tokens` (256 token) trong quá trình thực thi ghi đè, đoạn văn bản sinh ra khá dài và có thể lặp lại một số ý tưởng.

2.  **Tại sao các mô hình Decoder-only như GPT lại phù hợp cho tác vụ này?**
    Các mô hình Decoder-only như GPT (Generative Pre-trained Transformer) rất phù hợp cho tác vụ sinh văn bản (text generation) vì những lý do sau:

    *   **Kiến trúc tự hồi quy (Autoregressive Architecture):** Kiến trúc của các mô hình Decoder-only được thiết kế để dự đoán token tiếp theo dựa trên tất cả các token đã được sinh ra trước đó trong chuỗi. Điều này tạo ra một quá trình sinh văn bản tự nhiên, từng bước, rất giống với cách con người viết hoặc nói.
    *   **Tiền huấn luyện trên nhiệm vụ Mô hình Ngôn ngữ (Language Modeling):** GPT được tiền huấn luyện trên một lượng lớn dữ liệu văn bản với mục tiêu dự đoán từ tiếp theo trong một chuỗi. Quá trình này giúp mô hình học được ngữ pháp, cú pháp, phong cách và thông tin ngữ nghĩa rộng lớn của ngôn ngữ, cho phép nó tạo ra văn bản mạch lạc và tự nhiên.
    *   **Khả năng tạo văn bản dài và mạch lạc:** Nhờ kiến trúc tự hồi quy và khả năng hiểu ngữ cảnh rộng, các mô hình Decoder-only có thể tạo ra các đoạn văn bản dài, duy trì được sự mạch lạc và tính nhất quán về chủ đề trong suốt quá trình sinh.
    *   **Không có token `[MASK]`:** Khác với các mô hình Encoder-only cần token `[MASK]` để dự đoán từ bị thiếu, mô hình Decoder-only chỉ đơn giản là tiếp tục chuỗi văn bản đã cho, khiến chúng trở nên lý tưởng cho các tác vụ sinh văn bản mở đầu như viết truyện, bài luận, email, hoặc đoạn hội thoại.

````python
import torch
from transformers import AutoTokenizer, AutoModel

# 1. Chọn một mô hình BERT
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# 2. Câu đầu vào
sentences = ["This is a sample sentence."]

# 3. Tokenize câu
# padding=True: đệm các câu ngắn hơn để có cùng độ dài
# truncation=True: cắt các câu dài hơn
# return_tensors='pt': trả về kết quả dưới dạng PyTorch tensors
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# 4. Đưa qua mô hình để lấy hidden states
# torch.no_grad() để không tính toán gradient, tiết kiệm bộ nhớ
with torch.no_grad():
  outputs = model(**inputs)

# outputs.last_hidden_state chứa vector đầu ra của tất cả các token
last_hidden_state = outputs.last_hidden_state
# shape: (batch_size, sequence_length, hidden_size)

# 5. Thực hiện Mean Pooling
# Để tính trung bình chính xác, chúng ta cần bỏ qua các token đệm (padding tokens)
attention_mask = inputs['attention_mask']
mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
sum_embeddings = torch.sum(last_hidden_state * mask_expanded, 1)
sum_mask = torch.clamp(mask_expanded.sum(1), min=1e-9)
sentence_embedding = sum_embeddings / sum_mask

# 6. In kết quả
print("Vector biểu diễn của câu:")
print(sentence_embedding)
print("\nKích thước của vector:", sentence_embedding.shape)
````

1.  **Kích thước (chiều) của vector biểu diễn là bao nhiêu? Con số này tương ứng với tham số nào của mô hình BERT?**
    Dựa trên kết quả thực thi trước đó, kích thước của vector biểu diễn (vector nhúng câu) là `768` chiều (`torch.Size([1, 768])`).
    Con số `768` này tương ứng với tham số `hidden_size` (kích thước của không gian ẩn) của mô hình BERT đã được sử dụng (`bert-base-uncased`). Đây là kích thước của các vector biểu diễn đầu ra cho mỗi token từ lớp cuối cùng của mô hình BERT.

2.  **Tại sao chúng ta cần sử dụng `attention_mask` khi thực hiện Mean Pooling?**
    Chúng ta cần sử dụng `attention_mask` khi thực hiện Mean Pooling để đảm bảo rằng các token đệm (padding tokens) không ảnh hưởng đến việc tính toán vector trung bình của câu. Khi xử lý một lô (batch) các câu có độ dài khác nhau, các câu ngắn hơn sẽ được thêm các token đệm để tất cả các câu có cùng độ dài (để có thể xử lý song song bởi mô hình). Các token đệm này không mang thông tin ngữ nghĩa thực sự của câu.

    *   Nếu chúng ta tính trung bình tất cả các token (bao gồm cả token đệm) mà không dùng `attention_mask`, các giá trị của token đệm (thường là vector 0) sẽ làm sai lệch giá trị trung bình, khiến vector biểu diễn cuối cùng không phản ánh chính xác ngữ nghĩa của câu gốc.
    *   `attention_mask` là một tensor nhị phân (chứa 0 hoặc 1), trong đó 1 đánh dấu các token thực sự của câu và 0 đánh dấu các token đệm. Bằng cách nhân `last_hidden_state` (chứa các vector nhúng của tất cả các token) với `attention_mask` (sau khi mở rộng chiều), chúng ta sẽ làm cho các vector nhúng của token đệm trở thành 0. Sau đó, khi tính tổng và chia trung bình, chúng ta chỉ tính toán dựa trên các token thực sự của câu, loại bỏ ảnh hưởng của token đệm.
